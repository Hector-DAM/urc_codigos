```python
import numpy as np

# Datos de ejemplo
altura_planta_mt = [1.38, 1.4, 1.42, 1.54, 1.3, 1.55, 1.5, 1.6, 1.41, 1.34]
longitud_reticular_cm = [51, 60, 64, 73, 56, 75, 80, 76, 58, 70]
area_foliar = [4.8, 5.6, 5.8, 6.5, 5.3, 7.0, 8.1, 7.8, 5.9, 6.1]
pesa_pulpa = [115, 130, 138, 148, 122, 152, 160, 155, 135, 140]

# Construimos la matriz con los datos de los cafés
X = np.array([altura_planta_mt, longitud_reticular_cm, area_foliar, pesa_pulpa], dtype=np.float64).T


# Vamos a definir y calculas las matrices H, I y J

# La matriz I es la matriz identidad de 10x10
I = np.eye(10)

# La matriz J es la matriz de unos de 10x10
J = np.ones((10,10))

# La forma de encontrar la matriz H es H = I - (1/n) * J, vamos a calcularla
H = I - (1/10) * J

print(H)
print()
print(X)
```

## Ejercicio 1

```python
# Vamos a calcular la matriz H * X
HX = np.dot(H, X)

# Print the product matrix HX
print("Matrix H * X:")
print(HX)
```

## Ejercicio 2

```python
# Primero calculemos H transpuesta
Ht = H.T

# Ahora veamos si Ht y H son iguales

print(H)
print()
print(Ht,"\n")

if np.array_equal(H, Ht):
    print("H y H transpuesta son iguales")
else:
    print("H y H transpuesta son diferentes")
```

## Ejercicio 3

```python
# Veamos que H^2 es igual a H
# Primero calculemos H^2
H2 = np.dot(H, H)
print(H2)
print()
print(H,"\n")

# Comparamos H y H^2 permitiendo una pequeña tolerancia (1e-10 = 0.0000000001)
if np.allclose(H, H2, atol=1e-10):
    print("H y H^2 son prácticamente iguales dentro de la tolerancia especificada.")
else:
    print("H y H^2 son diferentes.")
```

## Ejercicio 4

```python
import numpy.linalg as la
# Vamos a calcular los valores propios y vectores propios de la matriz H

# Primero calculemos los valores propios de H
valores_propios = la.eigvals(H)
print(f"Los valores propios de H son: {valores_propios}")

# Ahora calculemos los vectores propios de H
vectores_propios = la.eig(H)[1]
print(f"Los vectores propios de H son: {vectores_propios}")
```
